```python
import numpy as np
import matplotlib.pyplot as plt

# Create a grid
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)

# Function 1: f1(x1, x2) = (x1 - x2)^2
Z1 = (X1 - X2)**2

# Function 2: f2(x1, x2) = x1^2 - 3*x2^2
Z2 = X1**2 - 3 * X2**2

# Plot for f1
plt.figure(figsize=(8, 6))
plt.contourf(X1, X2, Z1, levels=np.arange(0, 10, 1), cmap='hot')
plt.title('Level Sets of f1(x1, x2)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar(label='f1 value')

plt.tight_layout()
plt.show()

# Plot for f2
plt.figure(figsize=(8, 6))
plt.contourf(X1, X2, Z2, levels=np.arange(-10, 5, 1), cmap='plasma')
plt.title('Level Sets of f2(x1, x2)')
plt.xlabel('x1')
plt.ylabel('x2')
plt.colorbar(label='f2 value')

plt.tight_layout()
plt.show()

# List of x3 values from -10 to 10, selecting 13 evenly spaced values
x3_values = np.linspace(-10, 10, 13)

# Loop through x3 values
for x3 in x3_values:
    # Calculate Z3 for the current x3
    Z3 = X1**2 + 5 * X2**2 + 3 * x3**2
   
    # Create a single contour plot
    plt.figure(figsize=(8, 6))
    cf = plt.contourf(X1, X2, Z3, levels=np.linspace(0, 900, 13), cmap='magma')
    plt.contour(X1, X2, Z3, levels=np.linspace(0, 900, 13), colors='black', linewidths=0.5)
   
    # Labels and title
    plt.title(f'Level Sets of f3(x1, x2, x3) with x3 at {x3:.2f}')
    plt.xlabel('x1')
    plt.ylabel('x2')
    plt.colorbar(cf, label='f3 value')
   
    # Show
    plt.tight_layout()
    plt.show()
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a grid
x1 = np.linspace(-10, 10, 100)
x2 = np.linspace(-10, 10, 100)
X1, X2 = np.meshgrid(x1, x2)

# Function 1: f1(x1, x2) = (x1 - x2)^2 (extended to 3D with x3 = 0)
Z1 = (X1 - X2)**2

# Function 2: f2(x1, x2) = x1^2 - 3*x2^2 (extended to 3D with x3 = 0)
Z2 = X1**2 - 3 * X2**2

# Function 3: f3(x1, x2, x3) = x1^2 + 5*x2^2 + 3*x3^2
# For 3D, we'll use a 3D grid
x3 = np.linspace(-10, 10, 100)
X1_3D, X2_3D, X3_3D = np.meshgrid(x1, x2, x3, indexing='ij')
Z3 = X1_3D**2 + 5 * X2_3D**2 + 3 * X3_3D**2

# Create a figure with three subplots
fig = plt.figure(figsize=(15, 5))

# Plot for f1
ax1 = fig.add_subplot(131, projection='3d')
surf1 = ax1.plot_surface(X1, X2, Z1, cmap='hot', alpha=0.8)
ax1.contour(X1, X2, Z1, levels=np.arange(0, 10, 1), colors='black', offset=0)
ax1.set_title('Level Sets of f1(x1, x2) = (x1 - x2)^2')
ax1.set_xlabel('x1')
ax1.set_ylabel('x2')
ax1.set_zlabel('f1 value')
fig.colorbar(surf1, ax=ax1, label='f1 value')

# Plot for f2
ax2 = fig.add_subplot(132, projection='3d')
surf2 = ax2.plot_surface(X1, X2, Z2, cmap='plasma', alpha=0.8)
ax2.contour(X1, X2, Z2, levels=np.arange(-10, 5, 1), colors='black', offset=-30)
ax2.set_title('Level Sets of f2(x1, x2) = x1^2 - 3x2^2')
ax2.set_xlabel('x1')
ax2.set_ylabel('x2')
ax2.set_zlabel('f2 value')
fig.colorbar(surf2, ax=ax2, label='f2 value')

# Plot for f3 (using a 2D slice for visualization, e.g., x3 plane)
ax3 = fig.add_subplot(133, projection='3d')
# Take a 2D slice at x3 = 0 for simplicity in 3D plotting
Z3_slice = Z3[:, :, 50]  # Middle slice where x3  0
surf3 = ax3.plot_surface(X1, X2, Z3_slice, cmap='magma', alpha=0.8)
ax3.contour(X1, X2, Z3_slice, levels=np.arange(-6, 6, 1), colors='black', offset=0)
ax3.set_title('Level Sets of f3(x1, x2, x3) = x1^2 + 5x2^2 + 3x3^2 (x3=0 slice)')
ax3.set_xlabel('x1')
ax3.set_ylabel('x2')
ax3.set_zlabel('f3 value')
fig.colorbar(surf3, ax=ax3, label='f3 value')

# Adjust layout
plt.tight_layout()
plt.show()
```
